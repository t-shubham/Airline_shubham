```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data=pd.read_excel('Data_Train.xlsx')
```

```python
train_data.head()
```

```python
train_data.isna().sum()
```

```python
train_data.shape
```

```python
train_data.dropna(inplace=True)                 #inplac will update df at same time
```

```python
#train_data.dtypes
```

```python
train_data['Date_of_Journey']=pd.to_datetime(train_data['Date_of_Journey'])
train_data['Dep_Time']=pd.to_datetime(train_data['Dep_Time'])
train_data['Arrival_Time']=pd.to_datetime(train_data['Arrival_Time'])
```

```python
#train_data.dtypes
```

```python
train_data['Journey_day']=train_data['Date_of_Journey'].dt.day
train_data['Journey_month']=train_data['Date_of_Journey'].dt.month
```

```python
train_data.drop('Date_of_Journey', inplace=True,axis=1)
```

```python
# train_data.head()
```

```python
train_data['Arrival_hour']=train_data['Arrival_Time'].dt.hour
train_data['Arrival_minutes']=train_data['Arrival_Time'].dt.minute
```

```python
train_data['Dep_hour']=train_data['Dep_Time'].dt.hour
train_data['Dep_minutes']=train_data['Dep_Time'].dt.minute
```

```python
train_data.drop('Arrival_Time',inplace=True,axis=1)
train_data.drop('Dep_Time',inplace=True,axis=1)
#train_data.head()
```

```python
duration=list(train_data['Duration'])
```

```python
for i in range(len(duration)):
    if len(duration[i].split())==2:
        continue
    else:
        if 'h' in duration[i]:
            duration[i]= duration[i]+' 0m'
        else:
            duration[i]= '0h ' + duration[i]
```

```python
train_data['Duration']= duration
```

```python
def hour(x):
    return x.split()[0][0:-1]

def minute(x):
    return x.split()[1][0:-1]
```

```python
train_data['Duration_hour']=train_data['Duration'].apply(hour)
train_data['Duration_minute']=train_data['Duration'].apply(minute)
```

```python
train_data.drop('Duration',inplace=True,axis=1)
```

```python
#train_data.head()
```

```python
#train_data.dtypes
```

```python
train_data['Duration_hour']=train_data['Duration_hour'].astype(int)
train_data['Duration_minute']=train_data['Duration_minute'].astype(int)
```

```python
# categorical 
cat=[c for c in train_data.columns if train_data[c].dtype=='object']


#numerical
num=[n for n in train_data.columns if train_data[n].dtype!='object']
```

```python
#Nominal data dealed with Onehot
#ordinal data dealed with Label_encoding
```

```python
#train_data['Airline'].value_counts()
```

```python
categorical=train_data[cat]
```

```python
Airline=pd.get_dummies(categorical['Airline'],drop_first=True)
Source=pd.get_dummies(categorical['Source'],drop_first=True)
Destination=pd.get_dummies(categorical['Destination'],drop_first=True)
```

```python
Source.head()
```

```python
Destination.head()
```

```python
categorical['Route_1']=categorical['Route'].str.split('→').str[0]
categorical['Route_2']=categorical['Route'].str.split('→').str[1]
categorical['Route_3']=categorical['Route'].str.split('→').str[2]
categorical['Route_4']=categorical['Route'].str.split('→').str[3]
categorical['Route_5']=categorical['Route'].str.split('→').str[4]
```

```python
categorical.head()
```

```python
categorical.drop('Route',inplace=True,axis=1)
```

```python
categorical.isnull().sum()
```

```python
for i in ['Route_3','Route_4','Route_5']:
    categorical[i].fillna('None',inplace=True)
```

```python
for i in categorical.columns:
    print('{} has total {} categories'.format(i,len(categorical[i].value_counts())))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
e=LabelEncoder()
```

```python
for i in ['Route_1','Route_2','Route_3','Route_4','Route_5']:
    categorical[i]=e.fit_transform(categorical[i])
```

```python
categorical.head()
```

```python
categorical.drop('Additional_Info' ,axis=1,inplace=True)
```

```python
categorical['Total_Stops'].value_counts()
```

```python
d={'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4}
```

```python
categorical['Total_Stops']=categorical['Total_Stops'].map(d)
```

```python
categorical.head()
```

```python
data_train=pd.concat([categorical,Airline,Source,Destination,train_data[num]],axis=1)
```

```python
data_train.head()
```

```python
data_train.drop('Airline',axis=1,inplace=True)
data_train.drop('Source',axis=1,inplace=True)
data_train.drop('Destination',axis=1,inplace=True)
```

```python
#to see outliers

def plot(df,col):
    fig,(ax1,ax2)=plt.subplots(2,1)
    sns.distplot(df[col],ax=ax1)
    sns.boxplot(df[col],ax=ax2)
```
